```python
import spacy
from google.colab import files
import pandas as pd
import io
from PyPDF2 import PdfReader
```

```python

# Load the spaCy pre-trained model
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_text_from_pdf(pdf_file):
    reader = PdfReader(pdf_file)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text.strip()
```

```python
def extract_details_with_ner(text):
    doc = nlp(text)
    details = {"Name": None, "Email": None, "Phone": None}

    # Extract Name (PERSON entity)
    for ent in doc.ents:
        if ent.label_ == "PERSON" and not details["Name"]:
            details["Name"] = ent.text

    # Extract Email using regex (spaCy doesn't detect email natively)
    import re
    email = re.search(r'[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+', text)
    details["Email"] = email.group(0) if email else None

    # Extract Phone using regex
    phone = re.search(r'\b\d{10}\b|\+\d{1,2}\s?\d{10}|\(\d{3}\)\s?\d{3}-\d{4}', text)
    details["Phone"] = phone.group(0) if phone else None

    return details
```

```python
uploaded_files = files.upload()
```

```python
# Process files and extract details
summaries = []
for uploaded_file in uploaded_files.keys():
    print(f"Processing file: {uploaded_file}")
    try:
        # Read and process the uploaded PDF
        pdf_file = io.BytesIO(uploaded_files[uploaded_file])
        pdf_text = extract_text_from_pdf(pdf_file)

        # Extract details
        details = extract_details_with_ner(pdf_text)

        # Append results
        summaries.append({"file_name": uploaded_file, **details})
    except Exception as e:
        print(f"Error processing file {uploaded_file}: {e}")
```

```python
# Save the extracted details to CSV
summary_df = pd.DataFrame(summaries)
output_path = "NER_extracted_details.csv"
summary_df.to_csv(output_path, index=False)
files.download(output_path)
```
